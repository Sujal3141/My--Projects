```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
from nltk.corpus import stopwords
```

```python
with open('/content/Narnia.txt') as f:
    corpus = f.read()
```

```python
list_corpus=list(corpus)
```

```python
len(list_corpus)
```

```python
corpus = corpus.replace("\n","")
corpus = corpus.replace('\x0c',"")
```

```python
corpus
```

```python
list_corpus
```

```python
book = []


raw_sent = sent_tokenize(corpus)
for sent in raw_sent:
  book.append(simple_preprocess(sent))
```

```python
model = gensim.models.Word2Vec(window=10,min_count=2)
```

```python
model.build_vocab(book)
```

```python
model.train(book,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('edmund')
```
