```python
import nltk
nltk.download('punkt')

nltk.download('stopwords')
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
from nltk.corpus import stopwords
```

```python
with open('/content/Narnia.txt') as f:
    corpus = f.read()
```

```python
list_corpus=list(corpus)
```

```python
len(list_corpus)
```

```python
corpus = corpus.replace("\n","")
corpus = corpus.replace('\x0c',"")
```

```python
corpus
```

```python
def preprocess_text(text):
    segmented_words = []
    for word in text.split():
        segmented_words.extend(wordninja.split(word))
    return ' '.join(segmented_words)
```

```python
preprocessed_text = preprocess_text(corpus)
```

```python
preprocessed_text
def stopword_rem(preprocessed_text):
    stop_words = set(stopwords.words('english'))
    words = preprocessed_text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
preprocessed_text=stopword_rem(preprocessed_text)
```

```python
book = []


raw_sent = sent_tokenize(preprocessed_text)
for sent in raw_sent:
  book.append(simple_preprocess(sent))
```

```python
model = gensim.models.Word2Vec(window=10,min_count=2)
```

```python
model.build_vocab(book)
```

```python
model.train(book,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('man')
```
